```python
import rioxarray as rxr
import subprocess
```

Enter a title for the name of this Scoops run, as well as the name you want the input and output files to have.

```python
title = "Scoops3D example P; Seattle, non-homogeneous with dry conditions"

run_name = "P_seadry"
```

Next, We must specify the digital elevation map, or DEM, that you plan to use, as well as the output directory for your files.

```python
input = "input/seaclipDEM.asc"

output = "P_seadry_output/"
```

Set a groundwater pore pressure option

Choose from:
 - 'no': no pore pressure
 - 'ru': fraction of overlying column weight
 - 'pz': piezometric or water table surface

If not choosing 'no' or 'ru', also enter the unit weight of water ($\gamma_w$).

```python
gw = 'no'

gamw = 9.807
```

Set the number of layers used to define material properties for the slope-stability computations. **Do not add** more than that number of values to each list.

Cohesion: the cohesion $c$ for a layer\
phi: angle of internal friction $\phi$ for a layer in degrees\
gamt: total unit weight $\gamma_t$ for a layer of earth material

```python
lmat = 4
```

```python
cohesion = [209, 606, 0, 397]

phi = [38, 26, 34, 34]

gamt = [115, 108, 115, 115]
```

Set earthquake loading

eq: earthquake or seismic loading, imposed as a horizontal pseudo-acceleration coefficient ($k_{eq}$) and given as a dimensionless coefficient relative to a fraction of gravitational acceleration, $g$

```python
eq = 0
```

Set the limit-equilibrium method used for computing factor of safety.\
Specify method = ‘B’ to implement Bishop's simplified method or method = ‘O’ for the Ordinary (Fellenius) method

```python
method = 'B'
```

3D Search\
Choose from 'box', 'file', and 'single'.\
In the 'box' search approach, 4 sets of data are needed.

First, the bounding, in number of rows (`i_min, i_max`) and columns (`j_min, j_max`), of the topography file that you want to focus on.\ 
Additionally, a multiplier to control the nodal spacing of the search lattice is required. (`n_searches`)

Next, the minimum and maximum elevation (`z_min, z_max`), as well as the vertical spacing (`z_searches`) must be specified.

Finally, a few parameters to specify split directions and radius are required.
- dr: the radius increment used to construct trial surfaces
- deginc: the degree increment for analyzing additional slip directions (for a given trial surface) located to either side of the overall fall direction
- degmax: the maximum deviation, in degrees, to analyze slip directions on either side of the overall fall direction
- numdir: number of slip directions analyzed for each potential failure mass, including the overall fall direction $((2*degmax)/deginc + 1)$


```python
search = "box"

i_min, i_max = 1, 682
j_min, j_max = 1, 912
n_searches = 3

z_min, z_max = 50,1000

z_searches = 50

dr = 10

deginc = 0

degmax = 0

numdir = 1
```

### Box search parameters also require information on failures.

vacriterion: specifies whether volume or horizontal surface area is the primary criterion controlling potential failure size. ‘V’ indicates that the primary size criterion is volume; ‘A’ indicates that it is horizontal
surface area.

armin, armax: the minimum and maximum horizontal surface area limits for potential failure masses to be analyzed.

vmin, vmax: the minimum and maximum volume limits for potential failure masses to be analyzed.

tol: the volume or area tolerance level used when computing an initial potential failure size at each search-lattice node

limcol: the preferred minimum number of active DEM columns to be included in a potential failure
mass

```python
vacriterion = 'V'

armin, armax = 0,0

vmin, vmax = 106000, 1060000

tol = 10600

limcol = 100
```

Search parameters for a single trial surface a significantly simpler.\
The x, y, and z, coordinates should be specified, along with the radius and angle.\
The angle is specified as the azimuthal slip direction of the single trial surface in degrees counter-clockwise from the positive x-axis.

```python
xcen, ycen, zcen = 0,0,0

rad = 300

angle = 0
```

Finally, We must specify a few flags for output files with additional content.

isqout: flag to create three or four (for box search) search quality files. \
irelfos: flag to create an output file containing the relative factor of safety for each DEM cell.\
icritlattice: flag to create a 3D output file highlighting the critical nodes in the search lattice. \
ilattice: flag to create an output file containing the minimum factor of safety computed at each search-lattice node\
isubsurf: flag to create a 3D output file containing minimum factor-of-safety values in the subsurface. \
zfrac: fraction of the DEM cell size used to determine vertical spacing of the subsurface 3D factor-of-safety output file.

```python
isqout = 1

irelfos = 1

icritlattice = 1

ilattice = 1

isubsurf, zfrac = 0, 1
```

```python
conf = []
conf.append("title\n")
conf.append(f"{title}\n")
conf.append("lengthunits   ceeunits  gammaunits\n")
conf.append("m   kPa   kN/m^3\n") 
if gw == 'no':
    conf.append("water\n")
    conf.append(f"""{gw}\n""")
else:
    conf.append("water   gamw\n")
    conf.append(f"""{gw} {gamw}\n""")
conf.append("nmat\n")
conf.append(f"{lmat}\n")
conf.append("lnum   cee   phi   gamt\n")
for i in range(lmat):
    conf.append(f"""{i+1}   {cohesion[i]}.   {phi[i]}.   {gamt[i]}.\n""")
conf.append("eq\n")
conf.append(f"{eq}\n")
conf.append("method\n")
conf.append(f"{method}\n")
conf.append("srch\n")
conf.append(f"{search}\n")
conf.append("ismin   jsmin   ismax   jsmax   nsrchres\n")
conf.append(f"{i_min}   {j_min}   {i_max}   {j_max}   {n_searches}\n")
conf.append("zsmin   zsmax   zsrchres\n")
conf.append(f"{z_min}   {z_max}   {z_searches}\n")
conf.append("irefine   multres   fostol\n")
conf.append("1   8   0.01\n")
conf.append("dr   deginc   degmax   numdir\n")
conf.append(f"{dr}   {deginc}   {degmax}   {numdir}\n")   
conf.append("vacriterion   armin   armax   vmin   vmax   tol   limcol\n")
conf.append(f"{vacriterion}   {armin}   {armax}   {vmin}   {vmax}   {tol}   {limcol}\n")
conf.append("remove   foscut\n")
conf.append("M   5.0\n")
conf.append("isqout\n")
conf.append(f"{isqout}\n")
conf.append("irelfos\n")
conf.append(f"{irelfos}\n")
conf.append("icritlattice\n")
conf.append(f"{icritlattice}\n")
conf.append("ilattice\n")
conf.append(f"{ilattice}\n")
conf.append("isubsurf zfrac\n")
conf.append(f"{isubsurf}   {zfrac}\n")
conf.append("DEM file\n")
conf.append(f"{input}\n")
conf.append("layer file\n")
conf.append(f"input/sealayer\n")
conf.append("output directory\n")
conf.append(f"{output}\n")
input_file = f"{run_name}.scp"
f = open(input_file, "w")
f.write("".join(conf))
f.close()
```

```python
# subprocess.run(f"mkdir {output}", shell=True)
subprocess.run(f"./scoops {input_file}", shell=True)
```

Now that the program has run, we have a number of outputs to choose from.

1. The minimum 3D factor of safety for each cell
2. The volume or area (depending on the `vacriterion` variable, of the critical surface for each cell
3. The slope of the ground surface in each cell

Enter the corresponding number below to select the file to display.

```python
file_output = 2
```

```python
match file_output:
    case 1:
         out = rxr.open_rasterio(f"{output}/{run_name}_fos3d_out.asc",masked=True)
    case 2:
        if vacriterion in ['v','V']:
            out = rxr.open_rasterio(f"{output}/{run_name}_fosvol_out.asc",masked=True)
        if vacriterion in ['a','A']:
            out = rxr.open_rasterio(f"{output}/{run_name}_fosarea_out.asc",masked=True)
    case 3:
        out = rxr.open_rasterio(f"{output}/{run_name}_slope_out.asc",masked=True)
```

```python
out.plot()
```
