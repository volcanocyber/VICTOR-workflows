### PyFlowGo

PyFlowGo is an open-source platform for simulation of channelized lava thermo-rheological properties. It functions as an udpated version of FLOWGO, a 1-D model in which the velocity of a control volume flowing down a channel uses rheological properties computed following the lava cooling and crystallization path estimated via a heat balance box mod.

#### Attached is the newest version of the [manual](https://github.com/pyflowgo/pyflowgo/blob/master/PyFLOWGO_for_dummies.pdf) for additional parameters not included.

"Chevrel M.O., Labroquère J., Harris A. and Rowland S. (2018). PyFLOWGO: an open-source platform for simulation of channelized lava thermo-rheological properties. Computers and Geosciences Vol. 111. p.167–180. https://doi.org/10.1016/j.cageo.2017.11.009"

```python
import json
import subprocess
import pandas as pd
import matplotlib.pyplot as plt
import pprint
import plot_flowgo_results
```

### First, please input the name you will use to describe the simulation.

Additionally, add the path to the slope file (NOT a DEM), reference file provided for format. The step size is the $dt$ in meters in which to integrate.

```python
lava_name = "test"

slope_file = "./DEM_MaunaLoa1984.txt"

step_size = 10
```

### Heat Transfer Flags

Below are numerous flags to factor in additional aspects of heat transfer. 

If you would like to remove one heat budget from the existing set, simply switch "yes" to "no". We are currently working on ading the other heat budgets to this workflow.

```python
#choose from "basic", "2_emi", "lin_emi", "cont"
radiation = "basic"

# Input yes or no (all lower case) if you want to include the given type of flux
radiation_emissivity = "no"

conduction = "yes"

convection = "yes"

snyder = "no"

rain = "no"

viscous_heating = "no"
```

### Lava rheology parameters and models

The following models are required.

Choose from the set of calculation approaches described above each model.

```python
#Valid inputs for crystallization are: "basic" / "bimodal" / "bimodal_f_temp" / "melts" in units of dφ/dt 
crystallization_rate_model = "basic"

#Valid inputs for melt are: "basic" / "dragoni" / "shaw" / "vft" in units of η_melt
melt_viscosity_model= "vft"

#Valid inputs for relative viscosity are: "er" / "mp" / "kd" / "costa1" / "costa2" / "ptp1" / "ptp2" / "ptp3" in units of η_r
relative_viscosity_model = "er"

#Valid inputs include: "no" / "deformable"
relative_viscosity_bubbles_model = "no"

#Valid inputs for yield model are: "basic" / "dragoni" in units of τ_0
yield_strength_model = "basic"

#Valid inputs for the crust temperature model are: "basic" / "hon" / "bimodal" with units T_crust
crust_temperature_model = "constant"

#Valid inputs for effective cover model are: "basic" / "bimodal" with units f_crust
effective_cover_crust_model = "basic"

#Valid inputs for vesicle model are: "constant" / "bimodal" in units of φ_b
vesicle_fraction_model = "constant"

viscosity_eruption = 1000.0

max_packing = 0.641

einstein_coef = 3.27

strain_rate = 1.0
```

### Set initial channel dimensions and maximum length

Please set a width, depth, and maximum channel length to factor in these parameters into the heat equations.

```python
#Enter structural simulation parameters in meters
width = 4.5

depth = 1.4

max_channel_length = 30000
```

### Below, please enter parameters relating the temperature and heat transfer.

Temperatures, as well as other parameters that affect heat transfer and thermal interactions, are input below.

```python
#Set eruption temperature and viscosity
eruption_temperature = 1387.15

#Set temperature at the base of the flow and percentage of base layer over flow depth
basal_temperature = 773.15

core_base_distance = 19

#Set rainfall rate if rain is in the heat budget (m/s)
rainfall_rate = 7.93e-8

#Set wind speed (m/s), air temperature (K)
wind_speed = 5

air_temperature = 293.15

#CH value of air
ch_air = 0.0036
```

```python
#Difference between core temperature and hot component temperature
buffer = 140.0

#Percentage of crust covered
crust_cover_fraction = 1.0

#coefficient for velocity dependence of the crust cover
alpha = -.00756

crust_temperature = 773.15
```

### Please enter lava metadata

```python
#Starting Position of lava
position = 0.0

#Starting time
time = 1.0

#Percentage of lava that is crystallized
crystal_fraction = 0.104

#Density of the dense rock equivalent (DRE)
density_dre = 2970.0

#Percentage of the lava that is vesiclar
vesicle_fraction = 0.64

#Liquidus Temperature (K) of liquidus
liquidus_temperature = 0.0
```

```python
#Percentage of lava that turns into crystals during cooling
crystals_grown_during_cooling = 0.896

#Solid temperature of crystals
solid_temperature = 1237.15

#Latent heat of crystallization
latent_heat_of_crystallization = 350000.0
```

### Setup the JSON (input) file using the above parameters)

The next two cells properly format your parameters.

```python
lava_dict = {
  "lava_name": lava_name,
  "slope_file": slope_file,
  "step_size": step_size,
  "effusion_rate_init" : 3,

  "heat_budget_models":{
   "radiation" : radiation,
   "radiation_emissivity":radiation_emissivity,
   "conduction" : conduction,
   "convection" : convection,
   "rain" : rain,
   "snyder" : snyder,
   "viscous_heating" : viscous_heating
  },
  "models":{
    "crystallization_rate_model":crystallization_rate_model,
    "melt_viscosity_model":melt_viscosity_model,
    "relative_viscosity_model":relative_viscosity_model,
    "relative_viscosity_bubbles_model":relative_viscosity_bubbles_model,
    "yield_strength_model":yield_strength_model,
    "crust_temperature_model":crust_temperature_model,
    "effective_cover_crust_model": effective_cover_crust_model,
    "vesicle_fraction_model":vesicle_fraction_model
  },
  "terrain_conditions":{
    "width": width,
    "depth": depth,
    "gravity": 9.81,
    "slope_smoothing_active": False,
    "slope_smoothing_number_of_points": 10,
    "max_channel_length": max_channel_length
  },
  "eruption_condition":{
    "eruption_temperature" : eruption_temperature,
    "viscosity_eruption" : viscosity_eruption
  },
  "lava_state":{
    "position" : position,
    "time" : time,
    "crystal_fraction": crystal_fraction,
    "density_dre" : density_dre,
    "vesicle_fraction" : vesicle_fraction,
    "liquidus_temperature": liquidus_temperature
  },
  "radiation_parameters":{
    "stefan-boltzmann_sigma" : 5.67e-8,
    "emissivity_epsilon" : 0.95
  },
  "conduction_parameters":{
    "basal_temperature" : basal_temperature,
    "core_base_distance" : core_base_distance
  },
    "rain_parameters":{
    "rainfall_rate" : rainfall_rate,
    "density_water" : 958.0,
    "latent_heat_vaporization" : 2.8e6
  },
  "convection_parameters":{
    "wind_speed" : wind_speed,
    "ch_air" : ch_air,
    "air_temperature" : air_temperature,
    "air_density" : 0.4412,
    "air_specific_heat_capacity":1099.0
  },
  "thermal_parameters":{
    "buffer" : buffer,
    "crust_cover_fraction": crust_cover_fraction,
    "alpha" : alpha,
    "crust_temperature": crust_temperature
  },
  "melt_viscosity_parameters":{
    "shaw_slope" :2.36,
    "a_vft" : -4.52,
    "b_vft": 5558,
    "c_vft" : 582.9
  },
  "crystals_parameters":{
    "crystals_grown_during_cooling": crystals_grown_during_cooling,
    "solid_temperature": solid_temperature,
    "latent_heat_of_crystallization": latent_heat_of_crystallization
  },
  "relative_viscosity_parameters":{
    "max_packing": max_packing,
    "einstein_coef": einstein_coef,
    "strain_rate" : strain_rate
  }
}
```

```python
with open("template.json",'w') as f:
    json.dump(lava_dict, f)
```

### Run The Model

We suppress the output as pyFLOWGO sends output at every simulation step which leads to clutter in the notebook.

```python
subprocess.run("python main_flowgo.py template.json", shell=True, stdout=subprocess.DEVNULL)
```
