```python
import subprocess
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import rioxarray as rxr
from xrspatial import hillshade
import sys
```

### Enter date

First, please enter date-related parameters, and please choose a name for this simulation.

```python
year = 2005

month = 9

day = 7

hour = 5

minute = "00"

#Set duration of simulation in seconds
sim_duration = 1800

#If restart file exists, use it (NO/YES)
restart_run = "NO"

name = "twodee_test"
```

### Specify grid and topography parameters

Please specify grid spacing and origin in UTM, as well as if you are grabbing the topography from a file, or if you are creating a simple map with a slope,

```python
num_x = 400

num_y = 170

#grid spacing along x direction (m)
dx = 3

#grid spacing along y direction (m)
dy = 3

x_utm = 511300

y_utm = 4535600

utm_zone = "33S"

#Get the topography from a file
extract_topo_from_file = "YES"

#Set height of bottom left corner of topography, if not using file
z_origin = 0.0

#Slope of the flat topography in the x direction (degrees)
x_slope = 0.0

#Slope of the flat topography in the y direction (degrees)
y_slope = 0.0
```

### Input numerical terms

```python
#Set ambient gas density at 20°C (Kg/m^3)
ambient_gas_density = 1.204

#Set dense gas density at 20°C (Kg/m^3)
dense_gas_density = 1.839

averaged_temperature = 0.0

#Set exponent for dose calculation
dose_gas_toxic_exponent = 2
```

```python
front_froude_number = 1.0

optimal_courant_number = 0.25

edge_entrainment_coeff = 0.0

diffusion_coefficient = 0.20

shape_parameter = 0.5

zeta_parameter= 0.0

#Entrainment coefficient
alpha_2= 0.7

#Additional entrainment coefficient
alpha_3 = 1.3

#Final entrainment coefficient
alpha_7 = 0.45

von_karman_constant = 0.4

britter_b_constant = 0.1
```

### Enter meteorology terms

Please set the wind model and related coefficients/constants.

```python
#Set wind model (UNIFORM/DIAGNO), from file/diagno utility
wind_model = "UNIFORM"

#Set easting for the ground station
x_station = 511920

#Set northing for the ground station
y_station = 4535750

#Set reference height for temperature
z_ref = 3.0
```

### Enter file paths and output specifications

Please, if applicable, set paths for relevant model files as well as verboseness and inclusion of additional data in output files.

```python
#Only needed if extract_topo_from_file is set to "YES"
topography_file = "topography.grd"

#Set NONE/AAIG/GRD)
topo_file_format = "GRD"

#Only neeeded when roughness_model is set to "MATRIX"
roughness_file = "roughness.grd"

#Set NONE/AAIG/GRD)
roughness_file_format = "GRD"

#Only needed if run is a restart
restart_file = "restart.dat"

source_file = "source.dat"

#Can ignore if wind model is not DIAGNO
surf_data_file = "meteo_surface.dat"
   
#Can ignore if wind model is not MET_NC
met_nc_file = "wind.nc"

#Only needed when wind model is set to "SIMILARITY"
wind_file = "wind.dat"

#Only needed when wind model set to "DIAGNO"
diagno_file = "diagno.out"

#Directory that output files will be output to
output_directory = "outfiles"
```

### Output parameters

Please set which data to output.

```python
output_interval = 600

#Set output (yes/no). if yes, outputs results in grd format
output_grd_file_format = "no"                    

#Set output (yes/no). if yes, outputs results in netcdf format
output_nc_file_format  = "yes"

#Set output (yes/no). if yes, outputs topography
output_domain = "yes"

#Set output (yes/no). if yes, outputs roughness
output_z0 = "yes"                      

#Set output (yes/no). if yes, outputs source terms
output_source = "yes"

#Set output (yes/no). if yes, outputs meteo variables
output_meteo = "yes"

#Set output (yes/no). if yes, outputs velocity
output_velocity = "yes"

#Set output (yes/no). if yes, outputs averaged cloud thickness
output_h = "yes"

#Set output (yes/no). if yes, outputs averaged cloud density
output_rho = "yes"

#Set output (yes/no). if yes, outputs toxic dose
output_dose = "yes"

#Set output (yes/no). if yes, outputs concentration (ppm)
output_concentration = "yes"
```

### Generate input file and run model

Please run the next two cells to generate the Twodee input file and run the mode. Note that this will take multiple minutes to run.

```python
inp = f"""!-------------------------------
!
!  EXAMPLE OF TWODEE INPUT FILE
!
!-------------------------------

 TITLE
    PROBLEM_NAME = {name}
    !
 TIME
   YEAR = {year}
   MONTH = {month}
   DAY = {day}
   HOUR = {hour}
   MINUTE = {minute}
   SIMULATION_INTERVAL_(SEC) = {sim_duration}
   RESTART_RUN = {restart_run}
   !
 GRID
   UTM_ZONE = {utm_zone}
   UTM_DATUM = WGS_84
   !
   X_ORIGIN_(UTM_M) =  {x_utm}
   Y_ORIGIN_(UTM_M) = {y_utm}
   NX     = {num_x}
   NY     = {num_y}
   DX_(M) = {dx}
   DY_(M) = {dy}
   ! 
 PROPERTIES
   AMBIENT_GAS_DENSITY_20C_(KG/M3) = {ambient_gas_density}
   DENSE_GAS_DENSITY_20C_(KG/M3)   = {dense_gas_density}
   AVERAGED_TEMPERATURE_(C)        = {averaged_temperature}
   DOSE_GAS_TOXIC_EXPONENT         = {dose_gas_toxic_exponent}
   !
 METEO
   WIND_MODEL        = {wind_model}
   Z_REFERENCE_(M)   = {z_ref}
   !
   X_STATION_(UTM_M) = {x_station}
   Y_STATION_(UTM_M) = {y_station}
   !
   NC_DIM_X_NAME = lon
   NC_DIM_Y_NAME = lat
   NC_DIM_Z_NAME = z
   NC_DIM_T_NAME = time
   !
   NC_VAR_DATE_NAME = DATE
   NC_VAR_HOUR_NAME = HOUR
   NC_VAR_U_NAME    = U-VEL-WRF
   NC_VAR_V_NAME    = V-VEL-WRF
   NC_VAR_T_NAME    = T-WRF
    
 FILES
   OUTPUT_DIRECTORY        = {output_directory}
   !
   TOPOGRAPHY_FILE         = {topography_file}
   TOPOGRAPHY_FILE_FORMAT  = {topo_file_format}
   ROUGHNESS_FILE          = {roughness_file}
   ROUGHNESS_FILE_FORMAT   = {roughness_file_format}
   RESTART_FILE            = {restart_file}
   SOURCE_FILE             = {source_file}
   !
   WIND_FILE               = {wind_file}
   DIAGNO_FILE             = {diagno_file}
   SURF_DATA_FILE          = {surf_data_file}
   MET_NC_FILE             = {met_nc_file}
   !
   TRACK_POINTS            = NO
   TRACK_POINTS_FILE       = points.pts
   TRACK_BOXES             = NO
   TRACK_BOXES_FILE        = boxes.dat
   !
 OUTPUT
   OUTPUT_INTERVAL_(SEC)  = {output_interval}
   !
   OUTPUT_GRD_FILE_FORMAT = {output_grd_file_format}
   OUTPUT_NC_FILE_FORMAT  = {output_nc_file_format}
   !
   OUTPUT_DOMAIN          = {output_domain}
   OUTPUT_Z0              = {output_z0}
   OUTPUT_SOURCE          = {output_source}
   OUTPUT_METEO           = {output_meteo}
   !
   OUTPUT_VELOCITY        = {output_velocity}
   OUTPUT_H               = {output_h}
   OUTPUT_RHO             = {output_rho}
   OUTPUT_DOSE            = {output_dose}
   OUTPUT_CONCENTRATION   = {output_concentration}
   OUTPUT_MAX_CONCENTRATION = yes                    (Options:YES/NO  If YES, outputs maximum concentration in ppm)
     HEIGHTS_(M)          = 0.05 0.5 1.0 1.5 2.0 2.5 3.0              (Heights at which concentration is plot)
     CONCENTRATION_BG     = 350.                     (Background dense gas concentration in ppm)
   OUTPUT_Z_CRITICAL      = yes                      (Options:YES/NO  If YES, outputs z below which c>c_critical)
     CRITICAL_C_(%)       = 5 10 15                     (Critical concentration values in %)
   OUTPUT_IMPACT          = no                       (Options:YES/NO  If YES, outputs impact based on averagedconcentrations over 15,30 and 60 min)
   !
 NUMERIC
   FRONT_FROUDE_NUMBER     = {front_froude_number}
   OPTIMAL_COURANT_NUMBER  = {optimal_courant_number}
   EDGE_ENTRAINMENT_COEFF  = {edge_entrainment_coeff}
   DIFFUSION_COEFFICIENT   = {diffusion_coefficient}
   SHAPE_PARAMETER         = {shape_parameter}
   ZETA_PARAMETER          = {zeta_parameter}
   ALPHA_2                 = {alpha_2}
   ALPHA_3                 = {alpha_3}
   ALPHA_7                 = {alpha_7}
   VON_KARMAN_CONSTANT     = {von_karman_constant}
   BRITTER_B_CONSTANT      = {britter_b_constant}
    """
with open("twodee.inp",'w') as f:
    f.write(inp)
```

```python
subprocess.run("./twodee twodee.inp",shell=True, stdout=subprocess.DEVNULL)
```
